**Chapter 3 â€“ Classification**

Fetch the MNIST dataset

```python
import sklearn
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', as_frame=False)
```

Split data into test and train subsets with first 60,000 rows in the train subset

```python
import numpy as np

X, y = mnist.data, mnist.target


y = y.astype(np.int8)

X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

Create 1D arrays of boolean for train and test labels which is true when the target represents an odd number

```python
y_train_odd = (y_train % 2 == 1)
y_test_odd  = (y_test  % 2 == 1)
```

Train a SGDClassifier to identify the above class

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train_odd)
```

Show the confusion matrix of the classifier using 3-fold cross validation on the training set

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix

y_train_pred = cross_val_predict(sgd_clf, X_train, y_train_odd, cv=3)

cm = confusion_matrix(y_train_odd, y_train_pred)
print("Confusion Matrix (3-fold CV on training set):")
print(cm)
```

Also calculate its precision, recall, and F1-score

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y_train_odd, y_train_pred)
recall = recall_score(y_train_odd, y_train_pred)
f1 = f1_score(y_train_odd, y_train_pred)

print("\nMetrics (3-fold CV on training set):")
print(f"Precision: {precision:.3f}")
print(f"Recall:    {recall:.3f}")
print(f"F1-score:  {f1:.3f}")
```
